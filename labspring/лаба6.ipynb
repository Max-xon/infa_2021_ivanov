```python
import pandas as pd
df = pd.read_excel('task2_column_names.ods')

cols=df.columns
colnames=[]
for i in range(len(cols)):
    colnames.append(cols[i])

names=[]
for i in range (len(colnames)):
    
    result = ''

    if colnames[i][0] == 't':
        result='t'
        names.append(result)
        continue
        
    if colnames[i][0] == 'A':
        result = 'Контрольный образец,'
    elif colnames[i][0] == 'S':
        result = 'Тестируемый образец,'

    a=colnames[i][1:-1]
    result=result+' разбавлен в '+ a + ' раз'
        
    if colnames[i][-1] == '+':
        result = result + ' (Наличие модулятора)'
    elif colnames[i][-1] == '-':
        result = result + ' (Отсутствие модулятора)'
    
    names.append(result)

df = pd.read_excel('task2_column_names.ods', names=names, index_col=0)
filename = 'task2_renamed.xls'
df.to_excel(filename)
df
```
