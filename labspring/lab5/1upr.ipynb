```python
import pandas as pd

df1 = pd.read_csv('mutant.txt', sep='\t')

df1.plot(df1.columns[0], df1.columns[1], ylabel=df1.columns[1], title='mutant')

df2 = pd.read_csv('wild_type.txt', sep='\t')
df2.plot(df2.columns[0], df2.columns[1], ylabel=df2.columns[1], title='wild type')
```

```python
import pandas as pd

df = pd.read_csv('chromo.txt', sep='\t')
a=[]
for i in range (df.shape[0]):
    if df['Time (min)'][i]>=4 and df['Time (min)'][i]<=7:
        a.append(i)
df=pd.read_csv('chromo.txt', sep='\t', skiprows=min(a), skipfooter=df.shape[0]-max(a), names=['Time (min)', 'Value (mAU)'])
a,y=df.columns
df.plot(a,y, ylabel=df.columns[1], title = 'chromo (4-7 min)')
```
