```python
import matplotlib.pyplot as plt
import pandas as pd

students = pd.read_csv("StudentsPerformance.csv")

a = students[students["parental level of education"] == "bachelor's degree"] 
b = students[students["parental level of education"] == "some high school"]
c = students[students["parental level of education"] == "some college"] 
d = students[students["parental level of education"] == "master's degree"]
e = students[students["parental level of education"] == "high school"] 
f = students[students["parental level of education"] == "associate's degree"] 

for df in (a,b,c,d,e,f):
    plt.scatter(df["reading score"], df["math score"])
e_l_types = [ "associate's degree","bachelor's degree" "some college",  "high school","some high school", "master's degree"]
plt.legend(e_l_types)
plt.title('parental level of education,\n maths&reading scores')
plt.xlabel('reading score')
plt.ylabel('math score')
plt.show()

for df in (a,b,c,d,e,f):
    plt.scatter(df["reading score"], df["writing score"])
e_l_types = [ "associate's degree","bachelor's degree" "some college",  "high school","some high school", "master's degree"]
plt.legend(e_l_types)
plt.title('parental level of education,\n maths&reading scores')
plt.xlabel('reading score')
plt.ylabel('writing score')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

students = pd.read_csv("StudentsPerformance.csv")
parental_levelof_ed_types = ["bachelors\ndegree", "some\ncollege", 
                             'masters\ndegree', 'associates\ndegree', 'high\nschool', 'some\nhigh\nschool']
scores=["reading score", "math score", "writing score"]
a = students[students["parental level of education"] == "bachelor's degree"] 
b = students[students["parental level of education"] == "some high school"]
c = students[students["parental level of education"] == "some college"] 
d = students[students["parental level of education"] == "master's degree"]
e = students[students["parental level of education"] == "high school"] 
f = students[students["parental level of education"] == "associate's degree"]

for i in range (len(scores)):
    plt.boxplot((a[scores[i]], b[scores[i]], c[scores[i]], d[scores[i]], e[scores[i]], f[scores[i]]), labels=parental_levelof_ed_types)
    plt.title('Parental level of education,\n'+scores[i])
    plt.ylabel(scores[i])
    plt.show()

plt.show()
```
